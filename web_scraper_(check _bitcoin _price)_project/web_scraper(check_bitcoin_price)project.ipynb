```python
import requests
from bs4 import BeautifulSoup
import datetime
import csv
import pandas as pd

def check_price():
    URL = 'https://www.coindesk.com/price/bitcoin/'  # The address of the website from which we will retrieve data
    
    #User Agent carries the existing information of the user on the internet.
    # You don't need to know this, you can access it from the link below.
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36",
        "Accept-Encoding": "gzip, deflate",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
        "DNT": "1",
        "Connection": "close",
        "Upgrade-Insecure-Requests": "1"
    }

    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.content, "html.parser")

    title_element = soup.find("h1", class_="typography__StyledTypography-sc-owin6q-0 lerjFI long-title")
    if title_element:
        title = title_element.get_text(strip=True)
    else:
        title = "Title not found"

    price_element = soup.find("span", class_="currency-pricestyles__Price-sc-1v249sx-0 fcfNRE")
    if price_element:
        price = price_element.get_text(strip=True)
    else:
        price = "Price not found"

    now = datetime.datetime.now()

    # Writing data to CSV file
    with open('bitcoinprice.csv', 'a+', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        # Checking if the file is empty
        if f.tell() == 0:
            writer.writerow(['Title', 'Price', 'Date and Time'])  # Write the title line
        writer.writerow([title, price, now])  # Write a row of data

    # Reading CSV file with pandas
    df = pd.read_csv('bitcoinprice.csv')
    print(df)


check_price()
```

```python
# You can use the code below to make it work automatically.

import schedule
import time

# Calling check_price function every 1 houR
schedule.every().hour.do(check_price)

while True:
    schedule.run_pending()
    time.sleep(3600)  # 1 hour (3600 seconds) standby
```

```python
# If you want to try sending yourself an email (just for fun) when a price hits below a certain level, you can try it
# out with this script

import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

def send_mail():
    # SMTP server and port information
    smtp_server = 'smtp.gmail.com'
    smtp_port = 465

    # Sender email and password
    sender_email = 'info@sametyolcu.com'
    password = 'xxxxxxxxxxxxxx'

    # Receiver email
    receiver_email = 'receiver@example.com'

    # Email subject and content
    subject = "The Shirt you want is below $15! Now is your chance to buy!"
    body = "Samet, This is the moment we have been waiting for. Now is your chance to pick up the shirt of your dreams. Don't mess it up! Link here: https://www.amazon.com/Funny-Data-Systems-Business-Analyst/dp/B07FNW9FGJ/ref=sr_1_3?dchild=1&keywords=data+analyst+tshirt&qid=1626655184&sr=8-3"

    # Create email using MIMEText and MIMEMultipart
    message = MIMEMultipart()
    message['From'] = sender_email
    message['To'] = receiver_email
    message['Subject'] = subject
    message.attach(MIMEText(body, 'plain'))

    # Create SMTP connection and send email
    try:
        server = smtplib.SMTP_SSL(smtp_server, smtp_port)
        server.login(sender_email, password)
        server.sendmail(sender_email, receiver_email, message.as_string())
        print("Email sent successfully!")
    except Exception as e:
        print(f"Email sending error: {e}")
    finally:
        server.quit()

send_mail()
```
